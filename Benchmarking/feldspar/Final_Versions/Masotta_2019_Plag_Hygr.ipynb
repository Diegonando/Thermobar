## This notebook benchmarks to supporting spreadsheet of Masotta et al. (2019)
- It reads in from the spreadsheet "Masotta_2019_Benchmark.xlsx"

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import Thermobar as pt
pd.options.display.max_columns = None
```

```python
out_PL=pt.import_excel('Masotta_2019_Benchmark.xlsx', sheet_name="Thermobar_Input")
# This extracts a dataframe of all inputs
my_input_PL=out_PL['my_input']
# This extracts a dataframe of plag compositions from the dictionary "out"
Plags=out_PL['Plags']
Liqs_PL=out_PL['Liqs']
```

```python
Calc_H2O=pt.calculate_fspar_liq_hygr(liq_comps=Liqs_PL, plag_comps=Plags, equationH="H_Masotta2019",
                                   T=my_input_PL['T_C']+273.15, P=5)
```

```python
(my_input_PL['Calc_H2O_Mas']-Calc_H2O['H2O_calc']).iloc[0:2]
```

```python
plt.plot(my_input_PL['Calc_H2O_Mas'], Calc_H2O['H2O_calc'], 'ok')
plt.xlabel('Supporting Spreadsheet H2O')
plt.ylabel('Thermobar H2O')
```

```python
plt.plot(my_input_PL['Calc_H2O_Mas'], my_input_PL['Calc_H2O_Mas']-Calc_H2O['H2O_calc'], 'ok')
plt.xlabel('Supporting Spreadsheet H2O')
plt.ylabel('Absolute difference (Spreadsheet-Thermobar)')
```
