## This notebook compares the results from Thermobar to those calculated using the spreadsheet hosted on Keith Putirka's personal website
(http://www.fresnostate.edu/csm/ees/faculty-staff/putirka.html)

```python
import numpy as np
import pandas as pd
import sys
sys.path.append("../..")
import matplotlib.pyplot as plt
import Thermobar as pt
import sympy as sym
pd.options.display.max_columns = None
import inspect
from functools import partial
import warnings as w
w.filterwarnings("ignore", message="rubicon.objc.ctypes_patch has only been tested ")
w.filterwarnings("ignore", message="The handle")
```

```python
#my_input, myLabels, Experimental_PT1, myLiquids1, myCPXs1,  myOls1, myFspars1, myOPXs1=pw.import_excel_unknowns('2pYroxene.xlsx', sheetname="Sheet1")
out=pt.import_excel('2Pyroxene.xlsx', sheet_name="Sheet1")

my_input=out['my_input']
myCPXs1=out['Cpxs']
myOPXs1=out['Opxs']
myLiquids1=out['Liqs']
```

```python
# Checking inputs have read correctly
display(myCPXs1.head())
display(myOPXs1.head())
```

# First tests iterate an equation for pressure with an equation for temperature

## Testing iterating Equation 39 for pressure, Wood 1973 for temperature

```python
P_Put=my_input['SolvePT_Press_Eq39_Wood']
T_Put=my_input['SolvePT_Temp_Eq39_Wood'] # as Putirka spreasheet outputs in GPa
calc=pt.calculate_cpx_opx_press_temp(opx_comps=myOPXs1, cpx_comps=myCPXs1, equationP="P_Put2008_eq39", equationT="T_Wood1973")
P_py=calc['P_kbar_calc']
T_py=calc['T_K_calc']
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 10))
#plotting 1:1 line
ax1.plot([np.min(T_Put), np.max(T_Put)], [np.min(T_Put), np.max(T_Put)], '-r')
ax2.plot([np.min(P_Put), np.max(P_Put)], [np.min(P_Put), np.max(P_Put)], '-r')
ax1.plot(T_py, T_Put, 'ok')
ax1.set_xlabel('T Python (K)')
ax1.set_ylabel('T Put (K)')
ax2.plot(P_py, P_Put, 'or')
ax2.set_xlabel('P Python (kbars)')
ax2.set_ylabel('P Put (kbars)')
ax3.plot(T_py, T_py-T_Put, 'ok')
ax3.set_xlabel('T Python (K)')
ax3.set_ylabel('T Python - T Put')
ax4.plot(P_py, 1000*(P_py-P_Put), 'or')
ax4.set_xlabel('P Python (kbar)')
ax4.set_ylabel('P Python - P Put (bars)')
plt.subplots_adjust(left=0.125, bottom=None, right=0.9, top=None, wspace=0.5, hspace=None)
```

## eq 39 for pressure, and eq 37 for temperature (Both Putirka 2008)
- Some small discrepencies - excel outputs change as you press "enter". Likely results from iterative solving rather than anything to do with equations


```python
P_Put=my_input['SolvePT_Press_Eq37_Eq39']
T_Put=my_input['SolvePT_Temp_Eq37_Eq39'] # as Putirka spreasheet outputs in GPa
calc=pt.calculate_cpx_opx_press_temp(opx_comps=myOPXs1, cpx_comps=myCPXs1, equationP="P_Put2008_eq39", equationT="T_Put2008_eq37")
P_py=calc['P_kbar_calc']
T_py=calc['T_K_calc']
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 10))
#plotting 1:1 line
ax1.plot([np.min(T_Put), np.max(T_Put)], [np.min(T_Put), np.max(T_Put)], '-r')
ax2.plot([np.min(P_Put), np.max(P_Put)], [np.min(P_Put), np.max(P_Put)], '-r')
ax1.plot(T_py, T_Put, 'ok')
ax1.set_xlabel('T Python (K)')
ax1.set_ylabel('T Put (K)')
ax2.plot(P_py, P_Put, 'or')
ax2.set_xlabel('P Python (kbars)')
ax2.set_ylabel('P Put (kbars)')
ax3.plot(T_py, T_py-T_Put, 'ok')
ax3.set_xlabel('T Python (K)')
ax3.set_ylabel('T Python - T Put')
ax4.plot(P_py, 1000*(P_py-P_Put), 'or')
ax4.set_xlabel('P Python (kbar)')
ax4.set_ylabel('P Python - P Put (bars)')
plt.subplots_adjust(left=0.125, bottom=None, right=0.9, top=None, wspace=0.5, hspace=None)
```
