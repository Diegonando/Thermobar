## This notebook benchmarks to the equations of Wang et al. (2021) - https://ejm.copernicus.org/articles/33/621/2021/

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import Thermobar as pt
pd.options.display.max_columns = None
```

```python
out=pt.import_excel('Wang2021_SupportingCalculator.xlsx', sheet_name="Thermobar_Format")
my_input=out['my_input']
myCPXs1=out['Cpxs']
myLiquids1=out['Liqs']
```

## Testing equation 32dH

```python
Py=pt.calculate_cpx_only_temp(cpx_comps=myCPXs1, H2O_Liq=my_input['H2O_Liq'],
                                       equationT="T_Put2008_eq32dH", P=7)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))
ax1.plot(my_input['Eq. (32dH)']+273.15, Py, 'ok')
ax1.plot([1200, 1700], [1200, 1700], '-r')
ax1.set_xlabel('Calc P Spreadsheet')
ax1.set_ylabel('Calc P Thermobar')


ax2.plot(my_input['H2O_Liq'], my_input['Eq. (32dH)']+273.15-Py, 'ok')
ax2.set_xlabel('Calc T Spreadsheet')
ax2.set_ylabel('Calc T Spreadsheet - Thermobar')
fig.tight_layout()
```
