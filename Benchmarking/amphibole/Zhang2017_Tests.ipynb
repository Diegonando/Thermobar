```python
import numpy as np
import pandas as pd
import Thermobar as pt#
pd.options.display.max_columns = None
pd.options.display.max_rows = None
import matplotlib.pyplot as plt
```

```python
# This dataset contains a lot of amphiboles which throw up errors
import warnings as w
with w.catch_warnings():
    w.simplefilter('ignore')
    out=pt.import_excel('Zhang_Supp2_Testing.xlsx', sheet_name="For_Thermobar")
    my_input=out['my_input']
    myAmps1=out['Amps']
    myLiquids1=out['Liqs']
```

## Equation 1 
- Slight differences due to differences in molar masses on 2nd to 3rd decimal place. Well wtihin erorr. 

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))
Si_eq1=pt.calculate_amp_only_melt_comps(amp_comps=myAmps1, equation="Zhang17").SiO2_Eq1
ax1.plot(Si_eq1, my_input['wt% SiO2 eq1'], 'ok')
ax1.set_xlabel('Si Eq 1 Thermobar (wt%)')
ax1.set_ylabel('Si Eq 1 Supplement (wt%)')
                               
ax2.plot(Si_eq1, my_input['wt% SiO2 eq1']-Si_eq1, 'ok')
ax2.set_xlabel('Si Eq 1 Thermobar (wt%)')
ax2.set_ylabel('Si Suppelment-Thermobar (wt%)')
fig.tight_layout()
```
