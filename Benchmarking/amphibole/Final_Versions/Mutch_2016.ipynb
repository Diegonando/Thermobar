```python
import numpy as np
import pandas as pd
import Thermobar as pt#
pd.options.display.max_columns = None
import matplotlib.pyplot as plt
```

```python
# Test of our method vs. Euans Table 5 and spreadsheet
out3=pt.import_excel('Mutch_Amphibole.xlsx', sheet_name="Table5_transpoted")
my_input3=out3['my_input']
myAmps3=out3['Amps']
myLiquids3=out3['Liqs']
amp_calc=pt.calculate_amp_only_press(amp_comps=myAmps3, equationP="P_Mutch2016")
```

```python
# Discrepencies likely due to rounded molar masses
T_Put=my_input3['AlTot from spreadsheet'] 
T_py=amp_calc['Al_tot']
fig, ((ax1, ax2)) = plt.subplots(1, 2, figsize=(12, 5))
#plotting 1:1 line
ax1.plot([np.min(T_Put), np.max(T_Put)], [np.min(T_Put), np.max(T_Put)], '-r')
ax1.plot(T_py, T_Put, 'ok')
ax1.set_xlabel('Al(tot) Python')
ax1.set_ylabel('Al(tot) Mutch Spreadsheet')
ax2.plot(T_py, T_py-T_Put, 'ok')
ax2.set_ylabel('Al(tot) Python - AL(tot) Mutch')
ax2.set_xlabel('Al(tot) Python')
plt.subplots_adjust(left=0.125, bottom=None, right=0.9, top=None, wspace=0.5, hspace=None)
```
